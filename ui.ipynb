```python
#migrate database in postgreSQL to sqlite3

import sqlite3
import psycopg2
import pandas as pd

#connect to postgreSQL
conn = psycopg2.connect("dbname=postgres user=postgres password=postgres")
cur = conn.cursor()

#connect to sqlite3
conn2 = sqlite3.connect('db.sqlite3')
cur2 = conn2.cursor()

#query to get all tables
cur.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public'")
tables = cur.fetchall()

#loop through all tables
for table in tables:
    #query to get all columns
    cur.execute("SELECT column_name FROM information_schema.columns WHERE table_name = '" + table[0] + "'")
    columns = cur.fetchall()
    #query to get all data
    cur.execute("SELECT * FROM " + table[0])
    data = cur.fetchall()
    #create table in sqlite3
    cur2.execute("CREATE TABLE " + table[0] + " (" + ", ".join([i[0] for i in columns]) + ")")
    #insert data into sqlite3
    cur2.executemany("INSERT INTO " + table[0] + " VALUES (" + ", ".join(["?" for i in columns]) + ")", data)
    #commit changes
    conn2.commit()

#close connections
conn.close()
conn2.close()
```

```python
#klasördeki csv dosyalarını okuyup onlarla sqlite3 database'i oluştur

import sqlite3
import pandas as pd
import os

#connect to sqlite3
conn = sqlite3.connect('db.sqlite3')
cur = conn.cursor()

#loop through all csv files in directory
for file in os.listdir():
    if file.endswith(".csv"):
        #read csv file
        df = pd.read_csv(file)
        #create table in sqlite3
        df.to_sql(file[:-4], conn, if_exists='replace', index=False)

#close connection
conn.close()

#sqlite3 database'den veri çekip pandas dataframe'e dönüştür

import sqlite3
import pandas as pd

#connect to sqlite3
conn = sqlite3.connect('db.sqlite3')
cur = conn.cursor()

#query to get all tables
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cur.fetchall()

#show all tables
print(tables)    

#close connection
conn.close()
```

```python
conn = sqlite3.connect('db.sqlite3')
cur = conn.cursor()
def show_table(table_name):
    #query to get all data
    cur.execute("SELECT * FROM " + table_name)
    data = cur.fetchall()
    #show table
    print(*data)
    #commit changes
    conn.commit()
```

```python
# db.sqlite3 üzerinden calisan table'ındaki maaş ortalamasını bulan metodu yaz

import sqlite3
import pandas as pd

#connect to sqlite3

conn = sqlite3.connect('db.sqlite3')

cur = conn.cursor()

def average_salary():
    #query to get all data
    cur.execute("SELECT AVG(maas) FROM calisan")
    data = cur.fetchall()
    #show table
    print(*data)
    #commit changes
    conn.commit()

average_salary()
```

```python
# db.sqlite3 üzerinden calisan table'ındaki maaş column'unu dataframe'e çevir

import sqlite3
import pandas as pd

#connect to sqlite3

conn = sqlite3.connect('db.sqlite3')

cur = conn.cursor()

def salary_dataframe():
    #query to get all data
    cur.execute("SELECT maas FROM calisan")
    data = cur.fetchall()
    #show table
    df = pd.DataFrame(data)
    conn.commit()
    return df

maas_df = salary_dataframe()

```
